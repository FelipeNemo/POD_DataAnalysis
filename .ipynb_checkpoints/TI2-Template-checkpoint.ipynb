````python
# Criar matriz 10x10 preenchida com zeros
b = np.zeros((10, 10), dtype=float)

# Preencher a diagonal principal (índices iguais)
np.fill_diagonal(b, np.arange(10))

# Preencher a diagonal secundária (índices que somam 9)
indices = np.arange(10)
b[indices[1:], 9 - indices[1:]] = (indices[1:] * 11)

# Ajustar o valor 99 na base da diagonal secundária
b[9, 0] = 99

print("\nb:", b, sep="\n")
````

````python
# c) Multiplicar `a` e `b`
c = np.dot(a, b)
print("\nc:", c, sep="\n")
````

````python
# d) Calcular o vetor `d` (soma de cada linha de `c`)
d = c.sum(axis=1)

print("\nd:", d, sep="\n")
````

````python
# Criação do array 2D `a`
a = np.array([
    [85, 70, 27, 97, 52, 88, 84, 50,  7, 44],
    [62, 76,  9, 37, 22, 49, 72,  0, 26, 51],
    [19, 57, 73, 38, 60, 46, 34, 92, 32, 67],
    [95, 90, 64, 71,  3, 59, 11, 89, 47, 99],
    [17, 21, 10, 87, 43, 74, 86, 36,  4, 77],
    [68,  5, 45, 14, 29, 13, 20, 91, 35, 23],
    [58, 69, 18,  2, 75, 40, 82, 12, 98, 16],
    [66, 39, 54, 96,  6, 81, 94, 24, 28, 15],
    [30, 93, 56,  1, 48, 25, 33, 65, 61, 41],
    [83, 78, 80, 31, 55, 53, 42, 63, 79,  8]
])

# Impressão do array `a`
print("Array a:")
print(a)
````

````python
# Criação do array 1D `x`
x = np.array([30, 40, 50, 60, 70, 80, 90, 100, 110, 120])

# Impressão do array `x`
print("Array x:")
print(x)
````

````python
# Cálculo da distância euclidiana

# O vetor x é subtraído de cada linha do array a (broadcasting do NumPy).
# Calcula o quadrado de cada diferença.
# Soma os valores quadrados de cada linha.
# Aplica a raiz quadrada a cada soma.

distancias = np.sqrt(np.sum((a - x)**2, axis=1))
# Exibição do resultado
print("Distância euclidiana entre x e a:")
print(distancias)
````

````python
# Calcular as distâncias euclidianas
distances = np.linalg.norm(a - x, axis=1)

# Ordenar as distâncias e pegar os índices das 3 menores
indices = np.argsort(distances)[:3]

# Selecionar as 3 linhas mais próximas
k = a[indices]

print(k)
````

## 4. Funções de Ativação e Custo:
Dois grupos de função muito utilizadas em redes neurais são as funções de Atvação e as funções de Custo. As funções de ativação são utilizadas para introduzir não-linearidades nas redes neurais, enquanto as funções de custo são utilizadas para medir o quão bem o modelo está se saindo em relação aos dados de treinamento. Neste cenário, considere as seguintes funções:

**Funções de Ativação:**

Eq. Sigmoide:
$$ \sigma(x) = \frac{1}{1 + e^{-x}} $$

Eq. ReLU:
$$ ReLU(x) = \max(0, x) $$

Eq. Tangente Hiperbólica:
$$ tanh(x) = \frac{e^{x} - e^{-x}}{e^{x} + e^{-x}} $$

Eq.GeLU:
$$ GeLU(x) = 0.5x(1 + \tanh(\sqrt{2/\pi}(x + 0.044715x^3))) $$

Eq. Leaky ReLU:
$$ LeakyReLU(x) = \max(0.01x, x) $$

Eq. Swish:
$$ Swish(x) = \frac{x}{1 + e^{-x}} $$

**Funções de Custo:**

Eq. Entropia Cruzada:
$$ CE = - \sum_{i=1}^{n} y_i \log(\hat{y}_i) + (1 - y_i) \log(1 -\hat{y}_i)$$

**a. Crie um gráfico que inclua as seis funções de ativação mencionadas. Utilize 1000 valores de x indo de -5 a 5.**
**Saída esperada:**

<img src="arquivos/ativacoes.png">

````python
import numpy as np
import matplotlib.pyplot as plt

# Funções de ativação
def sigmoide(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

def gelu(x):
    return 0.5 * x * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * x**3)))

def leaky_relu(x, alpha=0.01):
    return np.maximum(alpha * x, x)

def swish(x):
    return x * sigmoide(x)

# Gerar valores de x
x = np.linspace(-5, 5, 1000)

# Lista das funções, títulos e cores
funcoes = [
    (sigmoide, "Sigmoide", "darkblue"),
    (relu, "ReLU", "red"),
    (leaky_relu, "Leaky ReLU", "green"),
    (tanh, "Tangente Hiperbólica", "yellow"),
    (gelu, "GeLU", "deeppink"),
    (swish, "Swish", "aquamarine")
]
# 2, 3, figsize=(20, 8)
# Criar layout de subplots (2 linhas x 3 colunas)
fig, axes = plt.subplots(2, 3, figsize=(14, 8))

# Iterar pelas funções, títulos, cores e subplots
for ax, (func, titulo, cor) in zip(axes.flat, funcoes):
    ax.plot(x, func(x), label=titulo, color=cor)
    #ax.set_title(titulo)
    #ax.set_xlabel("x")
    #ax.set_ylabel("y")
    ax.legend()
    ax.grid(True)

# Ajustar layout para melhor visualização
plt.tight_layout()
plt.show()
````

**b. Podemos dividir a função de custo da Entropia Cruzada em duas partes, como pode ser visto abaixo. Crie um gráfico que inclua as curvas dos dois lados da função. Crie um vetor `y` com 500 valores de 1, e logo depois crie um vetor `y_hat` com 500 valores que vão de 0.01 até 0.99.**

$$ CE = - \sum_{i=1}^{n} y_i \log(\hat{y}_i) + (1 - y_i) \log(1 -\hat{y}_i)$$
$$ PrimeiroTermo = - \log(\hat{y}_i)$$
$$ SegundoTermo = - \log(1 -\hat{y}_i)$$

**Saída esperada:**

<img src="arquivos/cross_entropy.png">

````python
import numpy as np
import matplotlib.pyplot as plt

# Criar vetor y e y_hat
y = np.ones(500)  # vetor y com 500 valores de 1
y_hat = np.linspace(0.01, 0.99, 500)  # vetor y_hat com valores de 0.01 até 0.99

# Calcular as duas partes da função de Entropia Cruzada
L1 = -np.log(y_hat)  # Parte 1: -log(y_hat) quando y = 1
L2 = -np.log(1 - y_hat)  # Parte 2: -log(1 - y_hat) quando y = 0

# Criar gráfico
plt.figure(figsize=(10, 6))

# Plotar as duas partes
plt.plot(y_hat, L2, label="y=1", color='red')
plt.plot(y_hat, L1, label="y=0", color='blue')


# Adicionar título e labels
plt.title("Cross Entropy Loss")
plt.xlabel("y_hat")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()
````

   - Use o critério do desvio padrão para detectar e remover outliers nas colunas `price` e `area`. Filtre os valores para manter apenas aqueles que estão dentro de três desvios padrão da média.



````python
from functools import reduce

def calculate_limits(df, column, num_std_dev=3):
    mean = df[column].mean()
    std = df[column].std()
    lower_limit = mean - (num_std_dev * std)
    upper_limit = mean + (num_std_dev * std)
    return lower_limit, upper_limit

def remove_outliers(df, columns, num_std_dev=3):
    # Usando map para calcular os limites de todas as colunas
    filters = map(
        lambda col: (df[col] >= calculate_limits(df, col, num_std_dev)[0]) & 
                    (df[col] <= calculate_limits(df, col, num_std_dev)[1]), 
        columns
    )
    # Reduzindo os filtros para combiná-los
    combined_filter = reduce(lambda x, y: x & y, filters)
    return df[combined_filter]

# Remover outliers das colunas 'price' e 'area'
df = remove_outliers(df, ['price', 'area'])

# Exibir informações após remoção de outliers
print("Shape após remoção de outliers:", df.shape)
````

5. **Normalização de Dados**  
   - Aplique a normalização z-score na coluna `area`, subtraindo a média e dividindo pelo desvio padrão. Isso permite que `area` seja analisada em uma escala padronizada.
   - A normalização z-score é dada pela fórmula:  
     $$ z = \frac{x - \mu}{\sigma} $$  
     onde `x` é o valor original, $\mu$ é a média e $\sigma$ é o desvio padrão.

````python
# Definir a função de normalização z-score para alterar a coluna diretamente
def normalize_z_score(df, column):
    mean = df[column].mean()
    std = df[column].std()
    df[column] = (df[column] - mean) / std

# Normalizar a coluna 'area' diretamente
normalize_z_score(df, 'area')
````

````python
# Definir a função de transformação de unidades para alterar a coluna diretamente
def transform_units(df, column, factor):
    df[column] = df[column] / factor

# Transformar 'price' para milhares diretamente
transform_units(df, 'price', 1000)
````

7. **Criação de Novas Colunas**  
   - Adicione uma nova coluna chamada `price_per_bedroom`, que representa o preço por quarto, dividindo `price` pelo número de quartos (`bedrooms`).


````python
def add_price_per_bedroom_column(df, price_column, bedrooms_column, new_column_name='price_per_bedroom', decimal_places=2):
    df[new_column_name] = df[price_column] / df[bedrooms_column].replace(0, float('nan'))
    df[new_column_name] = df[new_column_name].round(decimal_places)  # Limitar para 'decimal_places' casas decimais
    return df

# Adicionar a coluna 'price_per_bedroom'
df = add_price_per_bedroom_column(df, price_column='price', bedrooms_column='bedrooms')
print(df.shape)
````

## 7. Regressão Linear

**Crie um DataFrame que possua apenas as informações de área e preço do imóvel. Divida os dados em dois vetores, `X`(contendo a área) e `y` (contendo o preço). Adicione ao vetor X uma coluna de 1's para representar o termo independente. Em seguida, crie um vetor chamado `w` que contenha a seguinte expressão:**

### $$ w = (X^T X)^{-1} X^T y $$

**Em seguida, crie um gráfico de dispersão que mostre a relação entre a área e o preço do imóvel, inclua a reta que melhor se ajusta a esses dados. A reta é dada pela expressão:**

### $$ y = w_1 x + w_0 $$


**Dicas:**
- Utilize a função `np.linalg.pinv()` para calcular a pseudo-inversa de uma matriz.
- Garanta que `X.shape=(543,2)`, `y.shape=(543,)` e `w.shape=(2,)`.
- Para gerar a reta, você deve gerar um vetor de predições

**Saída esperada:**

<img src="arquivos/linear_regression.png">

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 2. Dividir os dados em dois vetores, X (área) e y (preço)
X = df['area'].values.reshape(-1, 1)  # Vetor de áreas
y = df['price'].values  # Vetor de preços

# 3. Adicionar uma coluna de 1's em X para representar o termo independente
X = np.hstack([np.ones((X.shape[0], 1)), X])

# 4. Calcular o vetor de pesos w
X_transpose = X.T
w = np.linalg.pinv(X_transpose @ X) @ X_transpose @ y  # Cálculo de w

# 5. Gerar a reta de predição
y_pred = X @ w  # Predições de y

# 6. Criar o gráfico de dispersão e a reta de regressão
plt.figure(figsize=(8, 6))

# Normalizando a coluna 'construction_year' para usar no mapeamento de cores
norm = plt.Normalize(df['construction_year'].min(), df['construction_year'].max())

# Criando o gráfico de dispersão com cores variando de roxo escuro a amarelo
scatter = plt.scatter(df['area'], df['price'], c=df['construction_year'], cmap='viridis', norm=norm, label='Dados reais')

# Adicionando a reta de regressão
plt.plot(df['area'], y_pred, color='red', label=f'Reta de regressão: $y = {w[1]:.2f}x + {w[0]:.2f}$')

plt.title('Gráfico de Dispersão de Área vs Preço Colorido pelo Ano de Construção')
plt.xlabel('Área (normalizada)')
plt.ylabel('Preço (em milhares)')
#plt.legend()
plt.grid(True)
plt.colorbar(scatter, label='Ano de Construção')  # Adiciona a barra de cores
plt.tight_layout()

# Exibir o gráfico
plt.show()
````
